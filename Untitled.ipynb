```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from textblob import TextBlob
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
data = pd.read_csv("clean_nus_sms.csv")
```

```python
text_data = data.drop(['Unnamed: 0', 'id', 'length', 'Date'], axis=1)
text_data.head()
```

```python
def data_processing(text):

    text = text.lower()
    text = re.sub(r"\@w+|\@w+|\#", '', text, flags=re.MULTILINE)
    text = re.sub(r"[^\w\s]", '', text)

    text_tokens = word_tokenize(text)
    filtered_text = [word for word in text_tokens if word not in stopwords.words('english')]

    return " ".join(filtered_text)
```

```python
text_data = text_data.astype(str)
text_data.Message = text_data['Message'].apply(data_processing)
text_data.Message = text_data.drop_duplicates('Message')
```

```python
lemmatizer = WordNetLemmatizer()

def lemmiting(data):
    text = [lemmatizer.lemmatize(word) for word in data]
    return data
```

```python
text_data['Message'] = text_data['Message'].astype(str)
text_data['Message'] = text_data['Message'].apply(lambda x: lemmiting(x))
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_data = text_data.astype(str)
text_data['polarity'] = text_data['Message'].apply(polarity)
```

```python
def sentiment(label):
    if label < 0:
        return 'Negative'
    elif label == 0:
        return 'Neutral'
    elif label > 0:
        return 'Positive'
```

```python
text_data['sentiment'] = text_data['polarity'].apply(sentiment)
```

```python
text_data.head()
```

```python
fig = plt.figure(figsize=(5,5))
sns.countplot(x='sentiment', data = text_data)
```

```python
fig = plt.figure(figsize=(7,7))
colors = ("yellowgreen", "gold", "red")
wp = {'linewidth':2, 'edgecolor':'black'}
tags = text_data['sentiment'].value_counts()
explode = (0.1, 0.1, 0.1)
tags.plot(kind='pie', autopct='%1.1f%%', shadow=True, colors=colors,
         startangle=90, wedgeprops=wp, explode=explode, label='')
plt.title('Distribiution of sentiments')
```
