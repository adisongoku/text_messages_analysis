```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from textblob import TextBlob
import re
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
import spacy
from top2vec import Top2Vec
```

```python
text_data = pd.read_csv("clean_nus_sms.csv")
```

```python
def data_processing(text):

    text = text.lower()
    text = re.sub(r"\@w+|\#", '', text, flags=re.MULTILINE)
    text = re.sub(r"[^\w\s]", '', text)
    text = re.sub(r"https?://\S+|www\.\S+", '', text)
    text = re.sub(r"<.*?>", '', text)

    text_tokens = word_tokenize(text)
    filtered_text = [word for word in text_tokens if word not in stopwords.words('english')]

    return " ".join(filtered_text)
```

```python
text_data.Message = text_data.Message.astype(str)
text_data.Message = text_data['Message'].apply(data_processing)
# text_data.Message = text_data.Message.drop_duplicates('Message', keep='first')
```

```python
lemmatizer = WordNetLemmatizer()

def lemmiting(data):
    text = [lemmatizer.lemmatize(word) for word in data]
    return data
```

```python
text_data['Message'] = text_data['Message'].astype(str)
text_data['Message'] = text_data['Message'].apply(lambda x: lemmiting(x))
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_data = text_data.astype(str)
text_data['polarity'] = text_data['Message'].apply(polarity)
```

```python
def sentiment(label):
    if label < 0:
        return 'Negative'
    elif label == 0:
        return 'Neutral'
    elif label > 0:
        return 'Positive'
```

```python
text_data['sentiment'] = text_data['polarity'].apply(sentiment)
```

```python
from nltk.corpus import twitter_samples
```

```python
pos_tweets = twitter_samples.strings('positive_tweets.json')
neg_tweets = twitter_samples.strings('negative_tweets.json')
```

```python
# sample positive tweets

print(pos_tweets[0])
print(pos_tweets[10])
print(pos_tweets[20])
print(pos_tweets[100])
```

```python
# sample negative tweets

print(neg_tweets[0])
print(neg_tweets[10])
print(neg_tweets[230])
print(neg_tweets[578])
```

```python
pos_tweets_tokens = twitter_samples.tokenized('positive_tweets.json')
neg_tweets_tokens = twitter_samples.tokenized('negative_tweets.json')
```

Preprocessing tweets tokens

```python
pos_clean_tweets_list = list()
neg_clean_tweets_list = list()
```

```python
for tweet in pos_tweets_tokens:
    for token in tweet:
        pos_clean_tweets_list.append(data_processing(token))
```

```python
for tweet in neg_tweets_tokens:
    for token in tweet:
        neg_clean_tweets_list.append(data_processing(token))
```

```python
print(len(pos_clean_tweets_list))
print(len(neg_clean_tweets_list))
```

```python
all_tweets_tokens = pos_clean_tweets_list + neg_clean_tweets_list

len(all_tweets_list)
```

```python
all_tweets_tokens = nltk.FreqDist(token for token in all_tweets_tokens)

print(all_tweets_tokens.most_common(10))

sentiment_features = [word for (word, freq) in all_tweets_tokens.most_common(10000)]

sentiment_features[:5]
```
